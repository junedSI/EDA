```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
Data = pd.read_csv("students_adaptability_level_online_education.csv")
Data.head(10)
```

```python
Data.columns
```

```python
Data.isnull()
```

```python
Data.isnull().sum()
```

```python
plt.figure(figsize=(20,10))
explode = [0.3,0.05,0.02]
labels = ['School', 'University', 'College']
colors = sns.color_palette("terrain_r", 7)
plt.pie(Data['Education Level'].value_counts(), labels=labels,
        colors = colors, autopct = '%0.0f%%',
        explode = explode,
        shadow = 'True',
        startangle = 90,
        )
plt.show()
```

```python
Data['Gender'].value_counts()
```

```python
plt.figure(figsize=(20,10))
explode = [0.3,0.05]
labels = ['Male', 'Female']
colors = sns.color_palette('Greys')
plt.pie(Data['Gender'].value_counts(), labels=labels,
        colors = colors, autopct = '%0.0f%%',
        explode = explode,
        shadow = 'True',
        startangle = 90,
        )
plt.show()
```

```python
Data["Age"].value_counts().plot(kind='bar',figsize=(20,10),color='green')
```

```python
Data['Financial Condition'].value_counts()
```

```python
plt.figure(figsize=(20,10))
explode = [0.3,0.05,0.02]
labels = ['Mid', 'Poor', 'Rich']
colors = sns.color_palette("terrain_r", 7)
plt.pie(Data['Financial Condition'].value_counts(), labels=labels,
        colors = colors, autopct = '%0.0f%%',
        explode = explode,
        shadow = 'True',
        startangle = 90,
        )
plt.show()
```

```python
sns.set_theme(font_scale=1.6) # set theme for plots
plt.figure(figsize=(30,10))
sns.barplot(data=Data,
                x="Education Level",
                y="Age", hue="Gender")

plt.title("bar plot")
plt.show()
```

```python
sns.set_theme(font_scale=1.6) # set theme for plots
plt.figure(figsize=(30,10))
sns.violinplot(data=Data,
               x="Education Level",
               y="Age", hue="Device"
               )

plt.title("bar plot")
plt.show()
```

```python
sns.set_theme(font_scale=1.6) # set theme for plots
plt.figure(figsize=(30,10))
sns.violinplot(data=Data,
               x="Education Level",
               y="Age", hue="Location"
               )

plt.title("bar plot")
plt.show()
```

```python
sns.set_theme(font_scale=1.6) # set theme for plots
plt.figure(figsize=(30,10))
sns.violinplot(data=Data,
               x="Education Level",
               y="Age", hue="Flexibility Level"
               )

plt.title("bar plot")
plt.show()
```

```python
plt.figure(figsize=(20, 10))

sns.lineplot(data = Data,
             x="Age",
             y="Education Level",
             hue = "Gender",
             palette = "muted",marker='D'
             )

plt.title("Line plot")
plt.show()
```
